<section>
  <h2>Modelo ARIMA</h2>
  <p>
    ARIMA, o AutoRegressive Integrated Moving Average, es una técnica estadística ampliamente utilizada para el análisis y pronóstico de series temporales. Su nombre deriva de tres componentes:
  </p>
  <ul>
    <li><strong>AR (AutoRegresivo):</strong> El modelo usa dependencias lineales entre una observación y un número de rezagos (valores anteriores) de la misma serie.</li>
    <li><strong>I (Integrado):</strong> Implica la diferenciación de los datos para hacer la serie estacionaria.</li>
    <li><strong>MA (Media Móvil):</strong> Modela el error de predicción como una combinación lineal de errores observados en el pasado.</li>
  </ul>
  <p>
    El modelo ARIMA se denota como ARIMA(p, d, q), donde <em>p</em> es el número de términos autorregresivos, <em>d</em> es el número de diferenciaciones necesarias para hacer estacionaria la serie, y <em>q</em> es el número de términos de media móvil.
  </p>
  <p>
    ARIMA es especialmente útil para series de datos univariadas, donde se busca comprender patrones pasados y predecir valores futuros, como la demanda mensual de productos o las ventas históricas.
  </p>
  <p>
    Estas técnicas son evaluadas con métricas como RMSE (Root Mean Squared Error), MAE (Mean Absolute Error) y R² (coeficiente de determinación), a fin de medir su precisión en la predicción de ventas.
  </p>
</section>
        
   </li>
    <li>

<section>
  <h2>Modelo de Regresión Lineal Múltiple</h2>
  <p>
    La regresión lineal múltiple es una extensión de la regresión lineal simple que permite predecir el valor de una variable dependiente continua a partir de dos o más variables independientes. El modelo se expresa matemáticamente como:
  </p>
  <p style="text-align:center; font-family: monospace; font-weight: bold;">
    Y = β₀ + β₁X₁ + β₂X₂ + ⋯ + βₙXₙ + ε
  </p>
  <p>
    Donde <strong>Y</strong> es la variable dependiente, <strong>X₁, X₂, ..., Xₙ</strong> son las variables independientes, <strong>β₀</strong> es el intercepto, <strong>β₁, β₂, ..., βₙ</strong> son los coeficientes de regresión que indican el cambio esperado en Y por unidad de cambio en cada X, y <strong>ε</strong> es el error aleatorio.
  </p>
  <p>
    Este modelo requiere cumplir ciertos supuestos: linealidad entre variables, normalidad y homocedasticidad de los residuos, ausencia de colinealidad entre variables independientes, y un número adecuado de observaciones (al menos 20 por variable). Se utiliza para entender la relación y el impacto relativo de múltiples variables predictoras sobre una variable de interés, y es ampliamente aplicado en predicción y análisis de tendencias.
  </p>
  <p>
    La interpretación de los coeficientes permite evaluar cómo cada variable independiente afecta la variable dependiente, manteniendo constantes las demás variables. Además, se evalúan estadísticas como el coeficiente de determinación (R²) para medir qué proporción de la variabilidad de Y es explicada por el modelo.
  </p>
</section>
        
   </li>
    <li>

<section>
  <h2>Modelo Random Forest</h2>
  <p>
    Random Forest es un algoritmo de aprendizaje automático basado en el ensamblaje de múltiples árboles de decisión para tareas de regresión y clasificación. En el contexto de regresión, Random Forest construye numerosos árboles de decisión entrenados con diferentes subconjuntos de datos y variables, y luego promedia sus predicciones para mejorar la precisión y reducir el sobreajuste.
  </p>
  <p>
    A diferencia de la regresión lineal múltiple, Random Forest no asume relaciones lineales ni supuestos estadísticos específicos, lo que le permite modelar relaciones complejas y no lineales entre variables. Es especialmente útil cuando existen interacciones complejas o variables con efectos no lineales sobre la variable objetivo.
  </p>
  <p>
    El modelo funciona generando árboles de decisión independientes mediante muestreo bootstrap y selección aleatoria de variables en cada división, lo que aporta diversidad y robustez. La predicción final es el promedio de las predicciones individuales de cada árbol.
  </p>
  <p>
    Random Forest proporciona además medidas de importancia de variables, permitiendo identificar cuáles predictores tienen mayor influencia en la variable objetivo. Es ampliamente utilizado en problemas de regresión cuando se busca un modelo flexible y con buen desempeño predictivo, aunque su interpretabilidad es menor comparada con la regresión lineal múltiple.
  </p>
</section>

<p>Convertir los IDs a string.</p>

```python
df_ventas["_id"] = df_ventas["_id"].astype(str)
df_ventas["cliente"] = df_ventas["cliente"].astype(str)
df_ventas["direccion"] = df_ventas["direccion"].astype(str)

df_cliente["_id"] = df_cliente["_id"].astype(str)

df_direccion["_id"] = df_direccion["_id"].astype(str)
df_direccion["cliente"] = df_direccion["cliente"].astype(str)

df_dventas["_id"] = df_dventas["_id"].astype(str)
df_dventas["producto"] = df_dventas["producto"].astype(str)
df_dventas["venta"] = df_dventas["venta"].astype(str)
df_dventas["cliente"] = df_dventas["cliente"].astype(str)

df_productos["_id"] = df_productos["_id"].astype(str)
```

<p>Hacer los Joins.</p>

```python
# Renombrar columnas en DVENTAS
df_dventas = df_dventas.rename(columns={"_id": "_id_dventas", "cliente": "cliente_dventas", "subtotal": "subtotal_dventas", "createdAt": "createdAt_dventas", "__v": "__v_dventas"})

# Renombrar columnas en CLIENTE
df_cliente = df_cliente.rename(columns={"_id": "_id_cliente", "telefono": "telefono_cliente", "dni": "dni_cliente", "createdAt": "createdAt_cliente", "__v": "__v_cliente", "pais": "pais_cliente"})

# Renombrar columnas en PRODUCTO
df_productos = df_productos.rename(columns={"_id": "_id_producto", "estado": "estado_producto", "createdAt": "createdAt_producto", "__v": "__v_producto"})

# Unir DVENTAS con CLIENTE
df_dventas_cliente = pd.merge(df_dventas, df_cliente, left_on="cliente_dventas", right_on="_id_cliente", how="left")

# Unir resultado con PRODUCTO
df_dventas_cliente_producto = pd.merge(df_dventas_cliente, df_productos, left_on="producto", right_on="_id_producto", how="left")

# VENTAS + DIRECCIÓN
# Renombrar columnas en VENTAS
df_ventas = df_ventas.rename(columns={"_id": "_id_ventas", "cliente": "cliente_ventas", "subtotal": "subtotal_ventas", "estado": "estado_ventas", "direccion": "direccion_ventas", "createdAt": "createdAt_ventas", "__v": "__v_ventas"})

# Filtrar dirección principal y renombrar
df_direccion_principal = df_direccion[df_direccion["principal"] == True].rename(columns={"_id": "_id_direccion", "dni": "dni_direccion", "telefono": "telefono_direccion", "direccion": "direccion_direccion", "pais": "pais_direccion", "createdAt": "createdAt_direccion", "__v": "__v_direccion"})

# Unir VENTAS con DIRECCIÓN
df_ventas_dir = pd.merge(df_ventas, df_direccion_principal, left_on="cliente_ventas", right_on="cliente", how="left")

# Unión final: (dventas + cliente + producto) + (ventas + direccion)
df_final = pd.merge(df_dventas_cliente_producto, df_ventas_dir, left_on="venta", right_on="_id_ventas", how="left")

print("✅ DataFrame final combinado:")
print(df_final.head())
```

<h2>4. ANÁLISIS EXPLORATORIO INICIAL (EDA)</h2>
<p>El análisis exploratorio inicial (EDA) es un método fundamental para comprender los datos mediante técnicas visuales y estadísticas antes de construir un modelo de regresión lineal múltiple.</p>
<p>En la regresión lineal múltiple, se evalúan las relaciones lineales entre la variable dependiente y múltiples variables independientes, verificando supuestos como la linealidad, normalidad y homocedasticidad de los residuos, ausencia de colinealidad, y la independencia de las muestras.</p>
<p>El EDA incluye el uso de gráficos de dispersión, matrices de gráficos de dispersión, histogramas, distribución normal, y análisis de coeficientes de determinación (R² y R² ajustado), así como la evaluación del error estándar residual para entender la precisión del modelo.</p>
<p>Este análisis permite identificar qué variables explicativas tienen mayor impacto y cómo se relacionan con la variable dependiente, facilitando la selección adecuada de variables para el modelo.</p>
<p>Además, el EDA ayuda a validar las suposiciones del modelo OLS (mínimos cuadrados ordinarios) y a comparar la eficacia de diferentes modelos antes de proceder al análisis confirmatorio.</p>

<h2>8. EVALUACION DEL MODELO</h2>

```python
def evaluar_modelo(y_real, y_pred, nombre_modelo):
    r2 = r2_score(y_real, y_pred)
    mae = mean_absolute_error(y_real, y_pred)
    mse = mean_squared_error(y_real, y_pred)
    rmse = np.sqrt(mse)
    
    return {
        'Modelo': nombre_modelo,
        'R2': r2,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse
    }
```
