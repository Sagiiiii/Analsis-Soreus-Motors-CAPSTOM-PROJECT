  <article id="modelo-arima">
    <h3>Modelo ARIMA</h3>
    <p>
      ARIMA (AutoRegressive Integrated Moving Average) es una técnica estadística ampliamente usada para analizar y pronosticar series de tiempo. Está compuesta por tres elementos:
    </p>
    <ul>
      <li><strong>AR (AutoRegresivo):</strong> Usa relaciones lineales entre una observación y sus valores pasados.</li>
      <li><strong>I (Integrado):</strong> Realiza diferenciaciones para convertir la serie en estacionaria.</li>
      <li><strong>MA (Media Móvil):</strong> Modela el error de predicción como combinación de errores anteriores.</li>
    </ul>
    <p>
      El modelo se representa como ARIMA(<em>p</em>, <em>d</em>, <em>q</em>), donde:
      <ul>
        <li><em>p</em>: número de términos autorregresivos</li>
        <li><em>d</em>: número de diferenciaciones</li>
        <li><em>q</em>: número de términos de media móvil</li>
      </ul>
    </p>
    <p>
      ARIMA es ideal para series univariadas como ventas o demanda mensual, y se evalúa con métricas como:
      <strong>RMSE</strong> (error cuadrático medio), <strong>MAE</strong> (error absoluto medio) y <strong>R²</strong> (coeficiente de determinación).
    </p>
  </article>

  <article id="modelo-rlm">
    <h3>Modelo de Regresión Lineal Múltiple</h3>
    <p>
      Este modelo predice una variable dependiente continua en función de dos o más variables independientes, representado por la fórmula:
    </p>
    <pre><code>Y = β₀ + β₁X₁ + β₂X₂ + ⋯ + βₙXₙ + ε</code></pre>
    <p>
      Donde:
      <ul>
        <li><strong>Y</strong>: variable dependiente</li>
        <li><strong>X₁...Xₙ</strong>: variables independientes</li>
        <li><strong>β</strong>: coeficientes del modelo</li>
        <li><strong>ε</strong>: error aleatorio</li>
      </ul>
    </p>
    <p>
      Se basa en supuestos como linealidad, homocedasticidad, normalidad de residuos y no colinealidad. Es útil para analizar tendencias y el impacto de múltiples factores sobre una variable objetivo.
    </p>
    <p>
      Las métricas clave para su evaluación son <strong>R²</strong>, <strong>RMSE</strong> y <strong>MAE</strong>.
    </p>
  </article>

  <article id="modelo-randomforest">
    <h3>Modelo Random Forest</h3>
    <p>
      Random Forest es un algoritmo de <em>machine learning</em> basado en conjuntos de árboles de decisión, aplicable tanto para regresión como para clasificación.
    </p>
    <p>
      En regresión, genera múltiples árboles sobre subconjuntos aleatorios de los datos y promedia sus predicciones, lo cual mejora la precisión y reduce el sobreajuste.
    </p>
    <p>
      A diferencia de RLM, no requiere cumplir supuestos estadísticos y puede modelar relaciones complejas o no lineales.
    </p>
    <p>
      También ofrece una <strong>importancia de variables</strong>, lo que permite identificar qué factores influyen más sobre la variable objetivo.
    </p>
    <p>
      Es una opción robusta y precisa, aunque menos interpretable que la regresión lineal.
    </p>
  </article>
</section>

```python
# Carga de colecciones a DataFrames
colecciones = {
    "ventas": "df_ventas",
    "clientes": "df_cliente",
    "direccions": "df_direccion",
    "dventas": "df_dventas",
    "productos": "df_productos",
    "inventarios": "df_inventarios"
}

for coleccion, df_name in colecciones.items():
    globals()[df_name] = pd.DataFrame(list(db[coleccion].find()))
```

```python
# Conversión de IDs y claves a string
for df in [df_ventas, df_cliente, df_direccion, df_dventas, df_productos, df_inventarios]:
    for col in df.columns:
        if col == "_id" or any(key in col for key in ["cliente", "producto", "venta", "direccion", "dventa", "inventario"]):
            df[col] = df[col].astype(str)
```

```python
# Renombrado de columnas
df_dventas.rename(columns={
    "_id": "_id_dventas", "cliente": "cliente_dventas", "producto": "producto_dventas",
    "venta": "venta_dventa", "cantidad": "cantidad_dventa", "subtotal": "subtotal_dventas",
    "createdAt": "createdAt_dventas", "__v": "__v_dventas"
}, inplace=True)

df_cliente.rename(columns={
    "_id": "_id_clientes", "telefono": "telefono_clientes", "dni": "dni_clientes",
    "createdAt": "createdAt_clientes", "pais": "pais_clientes", "__v": "__v_clientes"
}, inplace=True)

df_productos.rename(columns={
    "_id": "_id_productos", "estado": "estado_productos",
    "createdAt": "createdAt_productos", "__v": "__v_productos"
}, inplace=True)

df_inventarios.rename(columns={
    "_id": "_id_inventarios", "producto": "producto_inventarios",
    "cantidad": "cantidad_inventarios", "admin": "admin_inventarios",
    "createdAt": "createdAt_inventarios", "__v": "__v_inventarios"
}, inplace=True)

df_ventas.rename(columns={
    "_id": "_id_ventas", "cliente": "cliente_ventas", "subtotal": "subtotal_ventas",
    "estado": "estado_ventas", "direccion": "direccion_ventas",
    "createdAt": "createdAt_ventas", "__v": "__v_ventas"
}, inplace=True)
```

```python
# Filtro de direcciones principales y renombrado
df_direccion = df_direccion[df_direccion["principal"] == True].rename(columns={ 
    "_id": "_id_direccions", "cliente": "cliente_direccions",
    "dni": "dni_direccions", "telefono": "telefono_direccions",
    "direccion": "direccion_direccions", "pais": "pais_direccions",
    "createdAt": "createdAt_direccions", "__v": "__v_direccions"    
})
```

```python
# Uniones entre DataFrames
df_1 = pd.merge(df_dventas, df_cliente, left_on="cliente_dventas", right_on="_id_clientes", how="left")
df_2 = pd.merge(df_1, df_productos, left_on="producto_dventas", right_on="_id_productos", how="left")
df_3 = pd.merge(df_2, df_inventarios, left_on="_id_productos", right_on="producto_inventarios", how="left")
df_ventas_dir = pd.merge(df_ventas, df_direccion, left_on="cliente_ventas", right_on="cliente_direccions", how="left")
df_merge = pd.merge(df_3, df_ventas_dir, left_on="venta_dventa", right_on="_id_ventas", how="left")
```

```python
# Conversión de fechas y ordenamiento
if "createdAt_ventas" in df_merge.columns:
    df_merge["createdAt_ventas"] = df_merge["createdAt_ventas"].apply(
        lambda x: x["$date"] if isinstance(x, dict) and "$date" in x else x
    )
    df_merge["createdAt_ventas"] = pd.to_datetime(df_merge["createdAt_ventas"], errors="coerce")
    df_merge.sort_values("createdAt_ventas", inplace=True)
```

```python
# Eliminar columnas innecesarias
columnas_innecesarias = ['__v_dventas', '__v_clientes', '__v_productos', '__v_inventarios', '__v_ventas', '__v_direccions']
df_merge.drop(columns=[col for col in columnas_innecesarias if col in df_merge.columns], inplace=True)
```

```python
# Crear resumen por producto
df_resumen = df_merge.groupby('_id_productos').agg({
    'titulo': 'first',
    'stock': 'first',
    'nventas': 'first',
    'cantidad_inventarios': 'first'
}).reset_index()

df_resumen.rename(columns={
    'stock': 'stock_productos',
    'cantidad_inventarios': 'inventario'
}, inplace=True)
```

```python
# Convertir a numérico por si están como string
df_resumen['stock_productos'] = pd.to_numeric(df_resumen['stock_productos'], errors='coerce')
df_resumen['nventas'] = pd.to_numeric(df_resumen['nventas'], errors='coerce')

# Ahora sí se puede calcular el residuo
df_resumen['residuo'] = df_resumen['stock_productos'] - df_resumen['nventas']
```

```python
# Unir resumen al dataframe principal
df_merge = pd.merge(
    df_merge,
    df_ventas_reales,
    on='_id_productos',
    how='left'
)
df_merge = pd.merge(
    df_merge,
    df_resumen[['_id_productos', 'stock_productos', 'inventario', 'residuo']],    
    on='_id_productos',
    how='left'
)
```

```python
# Combinaciones a probar
p_values = [0, 1, 2]
d_values = [1]
q_values = [0, 1, 2]

resultados = []

for p in p_values:
    for d in d_values:
        for q in q_values:
            orden = (p, d, q)
            try:
                modelo = ARIMA(train_arima, order=orden)
                modelo_fit = modelo.fit()
                pred = modelo_fit.forecast(steps=dias_prueba)

                mae = mean_absolute_error(test_arima, pred)
                rmse = np.sqrt(mean_squared_error(test_arima, pred))
                r2 = r2_score(test_arima, pred)

                resultados.append({
                    'orden': orden,
                    'MAE': mae,
                    'RMSE': rmse,
                    'R2': r2
                })
            except Exception as e:
                print(f"❌ Falló para orden {orden}: {e}")

# Mostrar resultados ordenados por RMSE
df_resultados = pd.DataFrame(resultados).sort_values(by='RMSE')
df_resultados.head(10)  # Ver los 10 mejores
```
