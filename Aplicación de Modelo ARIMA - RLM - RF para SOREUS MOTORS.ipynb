  <article id="modelo-arima">
    <h3>Modelo ARIMA</h3>
    <p>
      ARIMA (AutoRegressive Integrated Moving Average) es una t√©cnica estad√≠stica ampliamente usada para analizar y pronosticar series de tiempo. Est√° compuesta por tres elementos:
    </p>
    <ul>
      <li><strong>AR (AutoRegresivo):</strong> Usa relaciones lineales entre una observaci√≥n y sus valores pasados.</li>
      <li><strong>I (Integrado):</strong> Realiza diferenciaciones para convertir la serie en estacionaria.</li>
      <li><strong>MA (Media M√≥vil):</strong> Modela el error de predicci√≥n como combinaci√≥n de errores anteriores.</li>
    </ul>
    <p>
      El modelo se representa como ARIMA(<em>p</em>, <em>d</em>, <em>q</em>), donde:
      <ul>
        <li><em>p</em>: n√∫mero de t√©rminos autorregresivos</li>
        <li><em>d</em>: n√∫mero de diferenciaciones</li>
        <li><em>q</em>: n√∫mero de t√©rminos de media m√≥vil</li>
      </ul>
    </p>
    <p>
      ARIMA es ideal para series univariadas como ventas o demanda mensual, y se eval√∫a con m√©tricas como:
      <strong>RMSE</strong> (error cuadr√°tico medio), <strong>MAE</strong> (error absoluto medio) y <strong>R¬≤</strong> (coeficiente de determinaci√≥n).
    </p>
  </article>

  <article id="modelo-rlm">
    <h3>Modelo de Regresi√≥n Lineal M√∫ltiple</h3>
    <p>
      Este modelo predice una variable dependiente continua en funci√≥n de dos o m√°s variables independientes, representado por la f√≥rmula:
    </p>
    <pre><code>Y = Œ≤‚ÇÄ + Œ≤‚ÇÅX‚ÇÅ + Œ≤‚ÇÇX‚ÇÇ + ‚ãØ + Œ≤‚ÇôX‚Çô + Œµ</code></pre>
    <p>
      Donde:
      <ul>
        <li><strong>Y</strong>: variable dependiente</li>
        <li><strong>X‚ÇÅ...X‚Çô</strong>: variables independientes</li>
        <li><strong>Œ≤</strong>: coeficientes del modelo</li>
        <li><strong>Œµ</strong>: error aleatorio</li>
      </ul>
    </p>
    <p>
      Se basa en supuestos como linealidad, homocedasticidad, normalidad de residuos y no colinealidad. Es √∫til para analizar tendencias y el impacto de m√∫ltiples factores sobre una variable objetivo.
    </p>
    <p>
      Las m√©tricas clave para su evaluaci√≥n son <strong>R¬≤</strong>, <strong>RMSE</strong> y <strong>MAE</strong>.
    </p>
  </article>

  <article id="modelo-randomforest">
    <h3>Modelo Random Forest</h3>
    <p>
      Random Forest es un algoritmo de <em>machine learning</em> basado en conjuntos de √°rboles de decisi√≥n, aplicable tanto para regresi√≥n como para clasificaci√≥n.
    </p>
    <p>
      En regresi√≥n, genera m√∫ltiples √°rboles sobre subconjuntos aleatorios de los datos y promedia sus predicciones, lo cual mejora la precisi√≥n y reduce el sobreajuste.
    </p>
    <p>
      A diferencia de RLM, no requiere cumplir supuestos estad√≠sticos y puede modelar relaciones complejas o no lineales.
    </p>
    <p>
      Tambi√©n ofrece una <strong>importancia de variables</strong>, lo que permite identificar qu√© factores influyen m√°s sobre la variable objetivo.
    </p>
    <p>
      Es una opci√≥n robusta y precisa, aunque menos interpretable que la regresi√≥n lineal.
    </p>
  </article>
</section>

```python
# Conexi√≥n a MongoDB
client = MongoClient("mongodb://localhost:27017")
db = client["tienda"]
```

```python
# Carga de colecciones a DataFrames
colecciones = {
    "ventas": "df_ventas",
    "clientes": "df_cliente",
    "direccions": "df_direccion",
    "dventas": "df_dventas",
    "productos": "df_productos",
    "inventarios": "df_inventarios"
}

for coleccion, df_name in colecciones.items():
    globals()[df_name] = pd.DataFrame(list(db[coleccion].find()))
```

```python
# Conversi√≥n de IDs y claves a string
for df in [df_ventas, df_cliente, df_direccion, df_dventas, df_productos, df_inventarios]:
    for col in df.columns:
        if col == "_id" or any(key in col for key in ["cliente", "producto", "venta", "direccion", "dventa", "inventario"]):
            df[col] = df[col].astype(str)
```

```python
# Renombrado de columnas
df_dventas.rename(columns={
    "_id": "_id_dventas", "cliente": "cliente_dventas", "producto": "producto_dventas",
    "venta": "venta_dventa", "cantidad": "cantidad_dventa", "subtotal": "subtotal_dventas",
    "createdAt": "createdAt_dventas", "__v": "__v_dventas"
}, inplace=True)

df_cliente.rename(columns={
    "_id": "_id_clientes", "telefono": "telefono_clientes", "dni": "dni_clientes",
    "createdAt": "createdAt_clientes", "pais": "pais_clientes", "__v": "__v_clientes"
}, inplace=True)

df_productos.rename(columns={
    "_id": "_id_productos", "estado": "estado_productos",
    "createdAt": "createdAt_productos", "__v": "__v_productos"
}, inplace=True)

df_inventarios.rename(columns={
    "_id": "_id_inventarios", "producto": "producto_inventarios",
    "cantidad": "cantidad_inventarios", "admin": "admin_inventarios",
    "createdAt": "createdAt_inventarios", "__v": "__v_inventarios"
}, inplace=True)

df_ventas.rename(columns={
    "_id": "_id_ventas", "cliente": "cliente_ventas", "subtotal": "subtotal_ventas",
    "estado": "estado_ventas", "direccion": "direccion_ventas",
    "createdAt": "createdAt_ventas", "__v": "__v_ventas"
}, inplace=True)
```

```python
# Filtro de direcciones principales y renombrado
df_direccion = df_direccion[df_direccion["principal"] == True].rename(columns={ 
    "_id": "_id_direccions", "cliente": "cliente_direccions",
    "dni": "dni_direccions", "telefono": "telefono_direccions",
    "direccion": "direccion_direccions", "pais": "pais_direccions",
    "createdAt": "createdAt_direccions", "__v": "__v_direccions"    
})
```

```python
# Uniones entre DataFrames
df_1 = pd.merge(df_dventas, df_cliente, left_on="cliente_dventas", right_on="_id_clientes", how="left")
df_2 = pd.merge(df_1, df_productos, left_on="producto_dventas", right_on="_id_productos", how="left")
df_3 = pd.merge(df_2, df_inventarios, left_on="_id_productos", right_on="producto_inventarios", how="left")
df_ventas_dir = pd.merge(df_ventas, df_direccion, left_on="cliente_ventas", right_on="cliente_direccions", how="left")
df_merge = pd.merge(df_3, df_ventas_dir, left_on="venta_dventa", right_on="_id_ventas", how="left")
```

```python
# Combinaciones a probar
p_values = [0, 1, 2]
d_values = [1]
q_values = [0, 1, 2]

resultados = []

for p in p_values:
    for d in d_values:
        for q in q_values:
            orden = (p, d, q)
            try:
                modelo = ARIMA(train_arima, order=orden)
                modelo_fit = modelo.fit()
                pred = modelo_fit.forecast(steps=dias_prueba)

                mae = mean_absolute_error(test_arima, pred)
                rmse = np.sqrt(mean_squared_error(test_arima, pred))
                r2 = r2_score(test_arima, pred)

                resultados.append({
                    'orden': orden,
                    'MAE': mae,
                    'RMSE': rmse,
                    'R2': r2
                })
            except Exception as e:
                print(f"‚ùå Fall√≥ para orden {orden}: {e}")

# Mostrar resultados ordenados por RMSE
df_resultados = pd.DataFrame(resultados).sort_values(by='RMSE')
df_resultados.head(10)  # Ver los 10 mejores
```

```python
# --- Preparar datos ---
dias_prueba = 90
train_arima = serie_sobrestock_diario[:-dias_prueba]
test_arima = serie_sobrestock_diario[-dias_prueba:]

# --- Entrenar modelo solo con datos anteriores ---
from statsmodels.tsa.arima.model import ARIMA

modelo_arima = ARIMA(train_arima, order=(2, 1, 1))  # Puedes ajustar a tu mejor orden
modelo_arima_fit = modelo_arima.fit()

# --- Predicci√≥n para los 90 d√≠as reservados ---
pred_arima = modelo_arima_fit.forecast(steps=dias_prueba)

# --- Evaluaci√≥n ---
mae_arima = mean_absolute_error(test_arima, pred_arima)
rmse_arima = np.sqrt(mean_squared_error(test_arima, pred_arima))
r2_arima = r2_score(test_arima, pred_arima)

print("=== Evaluaci√≥n del Modelo ARIMA (Backtesting 90 d√≠as) ===")
print(f"MAE: {mae_arima:.2f}, RMSE: {rmse_arima:.2f}, R¬≤: {r2_arima:.4f}")
```
